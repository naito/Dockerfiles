### 四則演算と優先順位

引き算や足し算、かけ算や割り算が混ざった複雑な計算をするときには、少し注意が必要です。かけ算や割り算が先に行われて、足し算や引き算はその後に行われ るからです。累乗はかけ算や割り算よりさらに先に計算されます。学校の勉強やテ ストで習ったのと同じく、四則演算には優先順位があります。

先ほど紹介した、4つの数字を使って10を作る計算式のうち、「`5/5+1*9`」という式を思い出してください。この式は、左から順に計算するのではなく、足し算の両辺にある割り算とかけ算が先に計算されます。その結果「`1+9`」となり、結果が10になるのです。

これも学校で習ったことと同じですが、丸括弧「`()`」を使うと、計算の順番をコントロールできます。丸括弧を使うと、あと何種類か10を計算する式を増やせます。

    5/5*(1+9)
    5/(5/(1+9))

### 変数名のつけ方

変数の名前をつけるには、簡単なルールがあります。プログラミング言語として決められているルールとしては、数字ではじまっていなければどんな文字列もPythonの変数名として使えるようになっています。

Python3では、日本語を使った変数名もつけられます。見た目は楽しいですが、なんのプログラムかよくわからないので、これは悪い変数名の例です。

    わん = 1
    にゃー = 2

好き勝手に変数名をつけると、プログラムが見にくくなってしまいます。そのため、Pythonではプログラムを作るときに守るべきルールがあります。次のようなルールを守ると、すっきりとした見やすいプログラムが書けます。

#### アルファベット、数字、アンダースコア(_)を組み合わせて作る

基本的にアルファベットの小文字を使って変数名を作りましよう。「`name1`」 「`name2`」のように数字を補うより、「`firstname`」「`lastname`」のように具体的な英語の 名詞や、「`sei`」「`mei`」のようなローマ字を使う方がわかりやすい変数名になります。

複数の単語を区切るときには「`some_word`」とアンダースコアを使うか、「`someWord`」のように単語の区切りに大文字を使います。どちらがよい、というルールはありません。見やすい、使いやすい方を選んで、ルールを統一すればよいと思います。

特別扱いしたい変数を定義するときにはアルファベットの大文字を使います。 Pythonには定数(中身を変更できない変数)がないため、大文字だけの変数名を持つ変数を定義して、定数のように扱うことがあります。

#### はじめの1文字はアルファベット

変数名の先頭に数字を使うことはできません。数値と区別がつかないからです。 Pythonでは、扱いに注意すべき特別な変数であることを示すために、アンダースコア からはじまる変数名をつけることがあります。書き換えては困る変数などが、アンダースコアではじめることが多いようです。

#### アルフアベットの大文字、小文字は区別される

Pythonでは「`girlsundpanzer`」と「`girlsUndPanzer`」は別の変数として扱われます。大文字と小文字を区別するからです。

#### 30個ほど、変数名に使えない単語がある

予約語と呼ばれる単語は変数名として使えません。変数として使おうとすると、エラー（SyntaxErfor、文法エラー）になります。例えば、「`and`」「`not`」「`if`」などは予約語して登録されているため、変数名として使えません。
